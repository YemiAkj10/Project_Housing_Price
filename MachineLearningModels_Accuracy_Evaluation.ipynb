```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
train_data = pd.read_csv("housing_train_clean.csv")
test_data = pd.read_csv("housing_test_clean.csv")
```

```python
Y_train = pd.DataFrame()
Y_train['SalePrice'] = train_data['SalePrice']
X_train = train_data.drop('SalePrice', axis=1)
X_train
```

```python
Y_test = pd.DataFrame()
Y_test['SalePrice'] = test_data['SalePrice']
X_test = test_data.drop('SalePrice', axis = 1)
X_test
```

```python
name_extra=[]
for name in X_train:
    if name not in X_test:
        name_extra.append(name)
        print(name_extra)
        name_extra
```

```python
X_train.shape
```

```python
X_train = X_train.values
Y_train = Y_train.values.reshape(-1)
X_test = X_test.values
Y_test = Y_test.values.reshape(-1)
```

```python
X_test.shape
```

```python
#MinMaxScaling X_train and X_test
min_X_train = np.min(X_train)
max_X_train = np.max(X_train)
X_train_Norm = (X_train - min_X_train) / (max_X_train - min_X_train)

print(X_train_Norm)

```

```python
min_X_test = np.min(X_test)
max_X_test = np.max(X_test)
X_test_Norm = (X_test - min_X_train) / (max_X_train - min_X_train)
print(X_test_Norm)
```

```python
def mape(Y_test, y_pred):
    Y_test, y_pred = Y_test, y_pred
    return np.mean(np.abs((Y_test - y_pred) / Y_test)) * 100
```

```python
KNR2 = KNeighborsRegressor()
param_grid = {'n_neighbors' : np.arange(1, 100, 10)}
KNR_gscv = GridSearchCV(KNR2, param_grid, cv = 5, scoring = 'neg_mean_absolute_percentage_error')
KNR_gscv.fit(X_train_Norm, Y_train)
print(KNR_gscv.best_params_)
```

```python
KNR = KNeighborsRegressor(n_neighbors = 11)
KNR.fit(X_train_Norm, Y_train)
y_pred = KNR.predict(X_test_Norm)
print("MSE:", metrics.mean_squared_error(Y_test, y_pred))
print("MAE:", metrics.mean_absolute_error(Y_test, y_pred))
print("MAPE:", mape(Y_test, y_pred))
```

```python
DTC = DecisionTreeRegressor()
param_dict = {'max_depth': np.arange(1,100,10)}
grid = GridSearchCV(DTC, param_grid = param_dict, cv = 5, scoring = 'neg_mean_absolute_percentage_error')
grid.fit(X_train, Y_train)
print(grid.best_params_)
```

```python
DTR = DecisionTreeRegressor(max_depth = 11)
DTR = DTR.fit(X_train_Norm, Y_train)
y_pred = DTR.predict(X_test_Norm)
print("MSE:", metrics.mean_squared_error(Y_test, y_pred))
print("MAE:", metrics.mean_absolute_error(Y_test, y_pred))
print("MAPE:", mape(Y_test, y_pred))
```

```python
RFC = RandomForestRegressor()
forest_params = {'max_depth' : np.arange(1, 100, 10)}
RFCV = GridSearchCV(RFC, forest_params, scoring = 'neg_mean_absolute_percentage_error', cv = 5)
RFCV.fit(X_train, Y_train)
print(RFCV.best_params_)
```

```python
RFR = RandomForestRegressor(max_depth = 71)
RFR.fit(X_train_Norm, Y_train)
y_pred = RFR.predict(X_test_Norm)
print("MSE:", metrics.mean_squared_error(Y_test, y_pred))
print("MAE:", metrics.mean_absolute_error(Y_test, y_pred))
print("MAPE:", mape(Y_test, y_pred))
```
